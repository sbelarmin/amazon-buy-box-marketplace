## 1. Business Context

Amazon marketplace sellers compete for the Buy Box, which determines the majority of product demand.

Sellers often assume that setting the lowest price is the optimal strategy to win the Buy Box and maximize sales.

However, lowering price reduces margins and may not always lead to higher overall profitability.

This creates a core strategic tension between:
- maximizing Buy Box win probability
- preserving contribution profit

The key question is whether aggressive price competition is economically optimal
