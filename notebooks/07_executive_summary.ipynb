# Amazon Buy Box Marketplace Analysis
## Executive Summary

---

## 1. Business Problem

Amazon marketplace sellers often assume they must offer the lowest price to win the Buy Box and maximize sales.

However, lower prices reduce contribution margin and may not maximize overall profit.

This project evaluates the tradeoff between pricing competitiveness and profitability in order to identify a price strategy that maximizes expected contribution profit.

---

## 2. Decision Question

What price should a seller set relative to competitors in order to maximize expected profit while remaining competitive for the Buy Box?

---

## 3. Analytical Approach

This analysis used a synthetic Amazon-style marketplace simulation with multiple sellers competing for the same ASIN over time.

The workflow included:

1. Marketplace exploration to understand competitive structure  
2. Descriptive modeling of Buy Box drivers  
3. Causal inference to isolate the effect of pricing on Buy Box outcomes  
4. Price optimization to identify the profit-maximizing price position  

---

## 4. Key Findings

### Buy Box dynamics
- The lowest priced seller wins the Buy Box most often, but not always  
- Fulfillment type, shipping speed, and seller rating also influence Buy Box outcomes  
- Buy Box ownership drives the majority of demand allocation  

### Causal findings
- Naive comparisons substantially overstated the effect of being the lowest priced seller  
- After adjusting for confounding, being the lowest priced seller still had a very large causal impact on Buy Box ownership  
- Regression adjustment and IPW produced similar results, indicating a robust causal relationship  

### Optimization findings
- Lower prices increase Buy Box probability but compress unit margin  
- Higher prices improve margin but reduce demand and Buy Box competitiveness  
- Expected profit is maximized at a price slightly below the lowest competitor  

---

## 5. Recommendation

Under the simulated marketplace conditions, the profit-maximizing strategy is to price approximately $2 below the lowest competitor.

At this point, the seller achieves approximately:
	•	~82% Buy Box win probability
	•	~$7.21 contribution margin per unit
	•	~$132 expected profit per seller-ASIN-day

This indicates that the optimal pricing strategy is not to aggressively undercut competitors, but to price modestly below them in order to balance conversion and margin.

However, this optimal price point is context-dependent and will vary based on:
	•	cost structure and margin constraints
	•	competitive intensity
	•	fulfillment advantages (e.g., FBA vs non-FBA)
	•	demand sensitivity to price

Therefore, the specific optimal price delta (e.g., ~$2 below the lowest competitor) should be interpreted as an illustrative result from this simulated environment, rather than a universal rule. In practice, sellers should continuously recalibrate pricing based on their own economics and competitive landscape.

---

## 6. Business Interpretation

These results suggest that sellers should not optimize for Buy Box share alone.

Instead, they should optimize for expected contribution profit by balancing:

- likelihood of winning the Buy Box  
- expected sales volume  
- contribution margin per unit  

In practical terms, pricing strategy should be treated as a margin-demand tradeoff rather than a race to the bottom.

---

## 7. Limitations

This project uses synthetic data and a simplified approximation of Amazon marketplace dynamics.

Limitations include:

- the true Amazon Buy Box algorithm is proprietary and more complex  
- competitors are not modeled as dynamically reacting to price changes  
- product cost is assumed to be constant across all sellers for a given ASIN
- FBA cost is a simplified per unit fulfillment cost and does not include storage fees, inbound shipping, etc.
- treatment effects are estimated under observed confounders only  

Despite these limitations, the project provides a realistic decision-science framework for evaluating marketplace pricing strategy.

---

## 8. Final Takeaway

Price is the dominant driver of Buy Box ownership, but lowest price is not always the most profitable strategy.

The strongest pricing strategy is one that balances competitiveness with margin preservation.

In this marketplace setting, the optimal seller behavior is to price **slightly below the lowest competitor**, rather than simply minimizing price.
