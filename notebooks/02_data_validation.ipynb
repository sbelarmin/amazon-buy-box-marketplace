## 2. Data Generation and Validation  

Objective:  

    1. generate and import synthetic marketplace data into dataframe  
    2. validate data matches intuition:   
      - Buy Box Winners sell more units on average  
      - Contribution Profit per Unit distribution is mostly positive  
      - Lowest Price sells more units on average  
---


```python
from src.data_generation.generate_marketplace_data import generate_marketplace_data
```

```python
# generate data experiment/synthetic dataset 
 
# generate data
products, sellers, seller_asin, df = generate_marketplace_data()

df.shape
```

```python
# validate buy box winner flag 
# winner should have a large units sold average
# loser should have a small units sold average

df.groupby("buy_box_winner")['units_sold'].mean()
```

```python
# reviewing the distribution of contribution profit per unit
# expected output should be distribution be mostly positive or profitable

df['contribution_profit_per_unit'].describe()
```

```python
# review if price rank 1 (lowest price) has the highest average units sold

df.groupby('price_rank')['units_sold'].mean()
```
